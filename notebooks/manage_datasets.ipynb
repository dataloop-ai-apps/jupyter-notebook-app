# Manage Datasets  
  
Datasets are buckets in the dataloop system that hold a collection of data items of any type, regardless of their storage location (on Dataloop storage or external cloud storage).  
  
## 1. Create Dataset  
  
To keep this notebook simple, we utilize the Dataloop storage and upload the files. However, Dataloop enables you to set up an integration with your cloud storage (AWS/GCP/Azure), create a storage driver to the specific buckets, and a dataset that index your data and enables data management over it. For more information on using your cloud storage programmatically, [read here](https://docs.dataloop.ai/docs/data-management-features-overview).


```python
import dtlpy as dl

project_id = 'project_id'
project = dl.projects.get(project_id=project_id)

dataset_name = 'dataset_name'

dataset = project.datasets.create(dataset_name=dataset_name)
```
